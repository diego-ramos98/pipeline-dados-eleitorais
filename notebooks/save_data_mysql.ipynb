***Conectando ao MYSQL***

```python
import mysql.connector

cnx = mysql.connector.connect(
    host= "localhost",
    user= "root",
    password = "123456"
)

print(cnx)
```

***Criando cursor para realizar as operações no mysql***

```python
cursor = cnx.cursor()
```

***Criando Banco de dados no mysql***

```python
cursor.execute("CREATE DATABASE IF NOT EXISTs dblocaisvotacoes;")
```

***CRIANDO TABELA***

*Vamos verificar os nomes da coluna do csv 

```python
import pandas as pd
```

```python
df_dados_eleitorais_rj = pd.read_csv("../data/dados_eleitorais_rj.csv")

df_dados_eleitorais_rj.head()
```

***CRIANDO TABELA***

```python
cursor.execute(""" 
        CREATE TABLE dblocaisvotacoes.tb_dados_eleitorais_rj(
            ID VARCHAR(255) PRIMARY KEY,
            LOCAL VARCHAR(255),
            MUNICIPIO VARCHAR(20),
            BAIRRO    VARCHAR(50),
            ENDERECO  VARCHAR(255),
            CEP       VARCHAR(20)
                      
               
        )


""")

```

***ADICIONANDO DADOS NA TABELA***

```python
list_data  = [tuple(row) for i, row in df_dados_eleitorais_rj.iterrows()]

list_data
```

```python
sql = "INSERT INTO dblocaisvotacoes.tb_dados_eleitorais_rj VALUES (%s, %s,%s,%s,%s,%s)"

cursor.executemany(sql,list_data)

```

```python
cnx.commit()
```

***VIZUALIZANDO OS DADOS***

```python
cursor.execute("SELECT * FROM dblocaisvotacoes.tb_dados_eleitorais_rj")

for row in cursor:
    print(row)
```
