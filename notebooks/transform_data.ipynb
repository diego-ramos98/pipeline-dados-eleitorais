***Conectando ao mongo***

```python
from pymongo.mongo_client import MongoClient

uri = "mongodb://localhost:27017/"

client = MongoClient(uri)

try:
    client.admin.command('ping')
    print("Conectou!!")
except Exception as e:
    print(e)
```

***LENDO OS DADOS***

```python
db = client["db_dados_eleitorais"]
collection = db["locais_votacoes"]
```

```python
for doc in collection.find():
    print(doc)
```

```python
lista_dados_eleitorais = []

for doc in collection.find():
    lista_dados_eleitorais.append(doc)
```

```python
import pandas as pd 
()
df_dados_eleitorais = pd.DataFrame(lista_dados_eleitorais)
```

***PEGANDO DADOS APENAS DO MUNICIPIO DO RIO***

```python
query = {"municipio" : "RIO DE JANEIRO"}

lista_dados_eleitorais_rj = []

for docs in collection.find(query):
    lista_dados_eleitorais_rj.append(docs)

lista_dados_eleitorais_rj
```

```python
len(lista_dados_eleitorais_rj)
```

***TRANSFORMANDO OS DADOS EM CSV***

```python
df_dados_eleitorais_rj = pd.DataFrame(lista_dados_eleitorais_rj)

df_dados_eleitorais_rj
```

```python
df_dados_eleitorais_rj = df_dados_eleitorais_rj.drop(["numLocal","numZona","secoes"],axis=1)
```

```python
df_dados_eleitorais_rj
```

```python
df_dados_eleitorais_rj.to_csv("../data/dados_eleitorais_rj.csv")
```
